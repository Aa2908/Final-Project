```python
# #login file
import hashlib

file_path = '/Users/aanyaparajuli/Desktop/task3/passwd.txt'

def hash_password(password):
    return hashlib.sha256(password.encode()).hexdigest()

# Prompt for login credentials
inp = input("Enter username: ")
passa = input("Enter Password: ")

# Hash the entered password for comparison
hashed_input_password = hash_password(passa)

# Read the existing content of the file
with open(file_path, 'r') as files:
    content = files.read()
    lines = content.splitlines()

# Check if the entered username exists
matching_lines = [line for line in lines if line.startswith(inp + ":")]

if matching_lines:
    # Extract real name and hashed password
    real_name, stored_hashed_password = matching_lines[0].split(":")[1:3]

    # Check if the entered password matches the stored hashed password
    if hashed_input_password == stored_hashed_password:
        print("Access granted.")
    else:
        print("Access denied.")
else:
    print("Access denied.")
```

```python
#deluser

import hashlib

file_path = '/Users/aanyaparajuli/Desktop/task3/passwd.txt'

def hash_password(password):
    return hashlib.sha256(password.encode()).hexdigest()

def delete_user(username, password):
    # Hash the entered password for comparison
    hashed_input_password = hash_password(password)

    # Read the existing content of the file
    with open(file_path, 'r') as files:
        content = files.read()
        lines = content.splitlines()

    # Check if the entered username exists
    matching_lines = [line for line in lines if line.startswith(username + ":")]

    if matching_lines:
        # Extract real name and hashed password
        real_name, stored_hashed_password = matching_lines[0].split(":")[1:3]

        # Check if the entered password matches the stored hashed password
        if hashed_input_password == stored_hashed_password:
            # Remove the matching line from the content
            lines.remove(matching_lines[0])

            # Write the updated content back to the file
            with open(file_path, 'w') as files:
                files.write('\n'.join(lines))

            print(f"User '{username}' deleted successfully.")
        else:
            print("Invalid password. User deletion failed.")
    else:
        print("User not found. User deletion failed.")

# Prompt for deletion credentials
del_user = input("Enter username to delete: ")
del_pass = input("Enter password for deletion: ")

# Call the delete_user function
delete_user(del_user, del_pass)
```

```python
import hashlib
from getpass import getpass

file_path = '/Users/aanyaparajuli/Desktop/task3/passwd.txt'

def hash_password(password):
    return hashlib.sha256(password.encode()).hexdigest()

def change_password(username, current_password, new_password):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    found = False
    for i in range(len(lines)):
        if lines[i].startswith(f"{username}:"):
            _, _, hashed_password = lines[i].strip().split(':')
            if hash_password(current_password) == hashed_password:
                # Password verification successful, update the password
                lines[i] = f"{username}:{new_password}\n"
                with open(file_path, 'w') as file:
                    file.writelines(lines)
                print("Password changed.")
                found = True
                break

    if not found:
        print("Username not found.")
    elif found and hash_password(current_password) != hashed_password:
        print("Current password is incorrect.")

# Input user information
usr_name = input("User: ")
current_password = getpass("Current Password: ")
new_password = getpass("New Password: ")
confirm_password = getpass("Confirm: ")

if new_password == confirm_password:
    change_password(usr_name, current_password, new_password)
else:
    print("New password and confirmation do not match.")
```
